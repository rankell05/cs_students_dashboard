```python
# Import pandas for data manipulation
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('cs_students.csv')

# Display first few rows to confirm loading
print("Dataset loaded successfully!")
df.head()
```

```python
# Count the number of records (rows) in the dataframe
num_records = len(df)
# Alternative: num_records = df.shape[0]

print(f"✔ Number of records (rows) in the dataset: {num_records}")
print(f"This means we have data for {num_records} students.")
```

```python
# Total elements = rows × columns
num_rows, num_cols = df.shape
total_elements = num_rows * num_cols

print(f"✔ Number of rows: {num_rows}")
print(f"✔ Number of columns: {num_cols}")
print(f"✔ Total elements (cells) in the dataset: {total_elements}")
print(f"This includes all student information across all columns.")
```

```python
# Get all column names
column_names = df.columns.tolist()

print("✔ Column names in the dataset:")
for i, col in enumerate(column_names, 1):
    print(f"   {i}. {col}")
```

```python
# Check data types of each column
column_types = df.dtypes

print("✔ Column data types:")
print(column_types)

print("\n" + "="*50 + "\n")
print("Summary of column types:")
print(df.dtypes.value_counts())
```

```python
# Get summary statistics for numeric columns only
print("✔ Summary statistics for numeric columns:")
print(df.describe())
```

```python
# Calculate standard deviation for numeric columns
# Select only numeric columns first
numeric_cols = df.select_dtypes(include=[np.number])

print("✔ Standard deviation for all numeric columns:")
print(numeric_cols.std())

print("\n" + "="*50 + "\n")
print("Explanation of standard deviation:")
print("- Measures how spread out the values are")
print("- Small std dev = values are close to the mean")
print("- Large std dev = values are spread out")
```

```python
# Get first 50 records using head()
first_50 = df.head(50)

# Calculate mean for numeric columns in first 50 records
print("✔ Mean values for the first 50 records:")
print(first_50.select_dtypes(include=[np.number]).mean())

print("\n" + "="*50 + "\n")
print("Note: Only numeric columns have mean values.")
print("Categorical columns (like Name, Gender) don't have meaningful means.")
```

```python
# GPA is the most critical column as it measures academic performance
print("="*60)
print("DETAILED ANALYSIS OF GPA COLUMN")
print("="*60)

# Find how many values in GPA column (use count method)
gpa_count = df['GPA'].count()
print(f"\n✔ Number of values in GPA column: {gpa_count}")

# Calculate the average GPA
avg_gpa = df['GPA'].mean()
print(f"✔ Average GPA: {avg_gpa:.2f}")

# Calculate basic statistics for GPA column
print(f"\n✔ Basic statistics for GPA column:")
print(f"   Count: {gpa_count}")
print(f"   Mean: {avg_gpa:.2f}")
print(f"   Median: {df['GPA'].median():.2f}")
print(f"   Mode: {df['GPA'].mode()[0]:.2f}")
print(f"   Standard Deviation: {df['GPA'].std():.2f}")
print(f"   Variance: {df['GPA'].var():.2f}")
print(f"   Minimum: {df['GPA'].min():.2f}")
print(f"   Maximum: {df['GPA'].max():.2f}")
print(f"   Range: {df['GPA'].max() - df['GPA'].min():.2f}")
print(f"   25th Percentile: {df['GPA'].quantile(0.25):.2f}")
print(f"   50th Percentile (Median): {df['GPA'].quantile(0.50):.2f}")
print(f"   75th Percentile: {df['GPA'].quantile(0.75):.2f}")
print(f"   90th Percentile: {df['GPA'].quantile(0.90):.2f}")

# Additional insights - FIXED: No nested f-strings
print(f"\n✔ Additional GPA Insights:")

# Calculate counts and percentages separately to avoid nested f-string issues
high_gpa_count = len(df[df['GPA'] >= 3.5])
high_gpa_percent = (high_gpa_count / len(df)) * 100

mid_gpa_count = len(df[(df['GPA'] >= 3.0) & (df['GPA'] < 3.5)])
mid_gpa_percent = (mid_gpa_count / len(df)) * 100

low_gpa_count = len(df[df['GPA'] < 3.0])
low_gpa_percent = (low_gpa_count / len(df)) * 100

print(f"   Students with GPA ≥ 3.5: {high_gpa_count} ({high_gpa_percent:.1f}%)")
print(f"   Students with GPA between 3.0 and 3.49: {mid_gpa_count} ({mid_gpa_percent:.1f}%)")
print(f"   Students with GPA < 3.0: {low_gpa_count} ({low_gpa_percent:.1f}%)")
```

```python
# Simple visualization to understand GPA distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(df['GPA'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(df['GPA'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {df["GPA"].mean():.2f}')
plt.axvline(df['GPA'].median(), color='green', linestyle='dashed', linewidth=2, label=f'Median: {df["GPA"].median():.2f}')
plt.title('GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.legend()

# Box plot
plt.subplot(1, 2, 2)
plt.boxplot(df['GPA'])
plt.title('GPA Box Plot')
plt.ylabel('GPA')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("The histogram shows how GPAs are distributed across students.")
print("The box plot shows the median, quartiles, and potential outliers.")
```

```python
print("="*60)
print("GPA COMPARISON ACROSS DIFFERENT GROUPS")
print("="*60)

# GPA by Gender
print(f"\n✔ Average GPA by Gender:")
print(df.groupby('Gender')['GPA'].agg(['mean', 'median', 'std', 'count']))

# GPA by Top Domains
print(f"\n✔ Average GPA by Interested Domain (Top 5):")
top_domains = df['Interested Domain'].value_counts().head(5).index
domain_gpa = df[df['Interested Domain'].isin(top_domains)].groupby('Interested Domain')['GPA'].mean().sort_values(ascending=False)
print(domain_gpa)

# GPA by Python Skill Level
print(f"\n✔ Average GPA by Python Skill Level:")
print(df.groupby('Python')['GPA'].mean())

# GPA by SQL Skill Level
print(f"\n✔ Average GPA by SQL Skill Level:")
print(df.groupby('SQL')['GPA'].mean())

# GPA by Java Skill Level
print(f"\n✔ Average GPA by Java Skill Level:")
print(df.groupby('Java')['GPA'].mean())
```
